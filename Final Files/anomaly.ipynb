```python
from pycaret.datasets import get_data
from pycaret.anomaly import *
anomaly = get_data('anomaly')
exp_ano = setup(anomaly)
```

```python
# we need to find the anomaly from this data set
anomaly
```

```python
## creating model 
#There are many types of models in th pycaret you can explore them in it's documentation, here we are using 'Minimum Covariance Determinant'.
# Now when we use the functoin below, it automatically takes the anomaly data as we have used the setup() function.
mcd = create_model('mcd')
```

```python
plot_model(mcd)
```

```python
knn = create_model('knn')
```

```python
plot_model(knn)
```

```python
pred = predict_model(mcd, data= anomaly)
pred
```
