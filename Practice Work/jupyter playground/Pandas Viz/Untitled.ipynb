```python
import numpy as np
import pandas as pd
%matplotlib inline
import cufflinks as cf
from plotly import __version__
from plotly .offline import download_plotlyjs,init_notebook_mode,plot,iplot
init_notebook_mode(connected=True)
cf.go_offline()
```

```python
df = pd.DataFrame(np.random.randn(100,4),columns='A B C D'.split())
df2= pd.DataFrame({'Category':['A','B','C'],'Values':[32,46,56]})
```

```python
df.iplot()
```

```python
df.iplot(kind='scatter',x='A',y='C',mode='markers',size=4)
```

```python
df2.iplot(kind='bar',x='Category',y='Values')
```

```python
df.sum().iplot(kind='bar') // count()// mean()
```

```python
df2.iplot(kind='box')
```

```python
df.iplot(kind='histogram',bins=30)
```

```python
df['C'].iplot(kind='hist',bins=24)
```

```python
df[['A','C','D']].iplot(kind='spread')
```

```python
df.iplot(kind='bubble',x='A',y='C',size='D')
```
